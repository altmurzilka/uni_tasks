```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
regression = pd.read_csv("AirPassengers.csv")
```

```python
#разделяем на тренировочную и тестовую части 
X = regression.iloc[:,1:-1].values
y = regression.iloc[:,-1].values
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = 0.25, random_state = 0)#разделяем данные на случайные подмножества
#random_state для постоянства значений
```

```python
#линейная регрессия
reg = LinearRegression()
reg.fit(xTrain, yTrain) #обучение модели
```

```python
#training set
plt.scatter(xTrain, yTrain, color = 'red')
plt.plot(xTrain, reg.predict(xTrain), color = 'purple')
plt.title('Value by Time (Training Set) ')
plt.xlabel(regression.columns[1])
plt.ylabel(regression.columns[-1])
plt.show()
```

```python
#test set
plt.scatter(xTest, yTest, color = 'red')
plt.plot(xTrain, reg.predict(xTrain), color = 'purple')
plt.title('Value by Time (Test set)')
plt.xlabel(regression.columns[1])
plt.ylabel(regression.columns[-1])
plt.show()
```

```python
yPred = reg.predict(xTest) #предсказывает вывод у
for i,j in zip(yTest, yPred):
    print('real ',i, ' - predicted: ', j)#сравнение полученных и предсказанных значений
```

```python
regression = pd.read_csv("bigcity.csv")

X = regression.iloc[:,1:-1].values
y = regression.iloc[:,-1].values

xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = 0.25, random_state = 0)

reg = LinearRegression()
reg.fit(xTrain, yTrain) 
    
plt.scatter(xTrain, yTrain, color = 'red')
plt.plot(xTrain, reg.predict(xTrain), color = 'purple')
plt.title('X by U (Training Set) ')
plt.xlabel(regression.columns[1])
plt.ylabel(regression.columns[-1])
plt.show()

plt.scatter(xTest, yTest, color = 'red')
plt.plot(xTrain, reg.predict(xTrain), color = 'purple')
plt.title('X by U (Test set)')
plt.xlabel(regression.columns[1])
plt.ylabel(regression.columns[-1])
plt.show()

yPred = reg.predict(xTest) 
for i,j in zip(yTest, yPred):
    print('real ',i, ' - predicted: ', j)
```
